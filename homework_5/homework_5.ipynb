Сигнатура функции:

```python

    mean_expressions = []

```

````python
def demonstrate_clt(expressions, sample_size, n_samples):
    mean_expressions = []
    for i in range(n_samples):
        sample_exp = np.random.choice(expressions, size=sample_size)
        mean_expressions.append(np.mean(sample_exp))
        
    return mean_expressions
````

````python
b_cells_clt = demonstrate_clt(b_cells_expression_data[example_gene], 1500, 500)
````

````python
sns.histplot(b_cells_clt, stat="density");
````

````python
nk_cells_clt = demonstrate_clt(nk_cells_expression_data[example_gene], 1500, 500)
````

А теперь посчитайте 95% доверительные интервалы для обоих распределений (примем тот факт, что средние распределены нормально для обоих клеточных типов) и скажите, отличается ли средняя экспрессия данного гена между клеточными типами?

````python
def calculate_ci(mean_expressions):
    mean = np.mean(mean_expressions)
    se = np.std(mean_expressions) / np.sqrt(len(mean_expressions))
    left_b = mean - 1.96 * se
    right_b = mean + 1.96 * se
    return left_b, right_b
````

````python
calculate_ci(b_cells_clt)
````

````python
calculate_ci(nk_cells_clt)
````

Средняя экспрессия отличается, так как границы доверительных интервалов для средних по выборкам не пересекаются.

Здесь же мы имеем дело с двумя выборками, поэтому будем проверять, пересекаются ли доверительные интервалы, и, если нет, то говорить о том, что средние в выборках отличаются.

````python
def check_intervals_intersect(first_ci, second_ci):
    if min(first_ci[1], second_ci[1]) - max(first_ci[0], second_ci[0]) <= 0:
        return False
    else:
        return True
````

Теперь для каждого гена посчитайте доверительные интервалы в обоих клеточных типах, и проверьте, пересекаются ли они? Результаты можно добавлять в список, например:

```python
ci_test_results = [True, False, ..., True]
```

Попытайтесь оформить это в виде функции, которая будет принимать на вход две таблицы с экспрессиями и выдавать для каждого гена, значимо ли отличается его средняя экспрессия между клеточными типами.

```python
def check_dge_with_ci(first_table, second_table):
    # dge - differential gene expression

    return ci_test_results
```

````python
def check_dge_with_ci(first_table, second_table):
    ci_test_results = {}
    for gene_name in first_table.iloc[:, :-1].columns:
        first_intervals = st.t.interval(alpha=0.95, 
                                        df=len(first_table[gene_name]) - 1, 
                                        loc=np.mean(first_table[gene_name]),
                                        scale=st.sem(first_table[gene_name]))
        second_intervals = st.t.interval(alpha=0.95, 
                                df=len(second_table[gene_name]) - 1, 
                                loc=np.mean(second_table[gene_name]),
                                scale=st.sem(second_table[gene_name]))
        ci_test_results[gene_name] = not check_intervals_intersect(first_intervals, second_intervals)
    
    return ci_test_results
````

````python
def check_dge_with_ztest(first_table, second_table):
    z_test_results = {}
    for gene_name in first_table.iloc[:, :-1].columns:
        z_result = ztest(first_table[gene_name], second_table[gene_name])
        z_test_results[gene_name] = z_result[1] < 0.05

    return z_test_results
````
